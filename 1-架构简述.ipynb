# Transformer模型的作用:
基于seq2seg架构的transformer模型可以完成NLP领域研究的典型任务,如机器翻译、文本生成等，同时又可以构建预训练语言模型，用于不同任务的迁移学习.
#### 一些设定
模拟使用Transformer模型架构处理从一种语言文本到另一种语言文本的翻译工作

命名遵循NLP中的规则：

Embeddding层将称作文本嵌入层

Embedding层产生的张量称为词嵌入张量

它的最后一维将称作词向量等
